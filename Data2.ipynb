```python
# python -m spacy download en_core_web_sm"

import spacy
import re
import pandas as pd
from tqdm import tqdm

# Charger spaCy pour l'anglais (une seule fois)
nlp = spacy.load("en_core_web_sm")

# Patterns regex pour détecter les données personnelles
EMAIL_PATTERN = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')
PHONE_PATTERN = re.compile(r'\b(?:\+?1[-.\s]?)?\(?[0-9]{3}\)?[-.\s]?[0-9]{3}[-.\s]?[0-9]{4}\b')

def detect_personal_data_rgpd(text):
    """
    Détecte toutes les données personnelles pour la conformité RGPD
    """
    if not text or pd.isna(text):
        return {
            'has_personal_data': False,
            'names': [],
            'emails': [],
            'phones': [],
            'addresses': []
        }
    
    text_str = str(text)
    
    # 1. EMAILS
    emails = EMAIL_PATTERN.findall(text_str)
    
    # 2. TÉLÉPHONES  
    phones = PHONE_PATTERN.findall(text_str)
    
    # 3. NOMS avec spaCy NER
    doc = nlp(text_str)
    names = []
    for ent in doc.ents:
        if ent.label_ == 'PERSON':
            names.append(ent.text)
    
    return {
        'has_personal_data': bool(names or emails or phones),
        'names': names,
        'emails': emails,
        'phones': phones,
    }

def anonymize_text_rgpd(text, detection_result):
    """
    Anonymise automatiquement le texte pour la conformité RGPD
    """
    if not detection_result['has_personal_data']:
        return text
    
    anonymized_text = str(text)
    
    # Anonymiser dans l'ordre : emails, téléphones, adresses, noms
    for email in detection_result['emails']:
        anonymized_text = anonymized_text.replace(email, "[EMAIL]")
    
    for phone in detection_result['phones']:
        anonymized_text = anonymized_text.replace(phone, "[PHONE]")
    
    for name in detection_result['names']:
        anonymized_text = anonymized_text.replace(name, "[NAME]")
    
    return anonymized_text
```

```python
# ÉTAPE 1: Configuration du nettoyage de données
import re
import string
import pandas as pd
from tqdm import tqdm

def clean_text(text):
    """
    Fonction complète de nettoyage de texte pour le NLP
    """
    if not text or pd.isna(text):
        return ""
    
    # Convertir en string au cas où
    text = str(text)
    
    # 1. CASSE : Convertir en minuscules
    text = text.lower()
    
    # 2. EMOJIS : Supprimer les emojis (patterns Unicode)
    emoji_pattern = re.compile("["
                              u"\U0001F600-\U0001F64F"  # emoticons
                              u"\U0001F300-\U0001F5FF"  # symboles & pictogrammes
                              u"\U0001F680-\U0001F6FF"  # transport & cartes
                              u"\U0001F1E0-\U0001F1FF"  # drapeaux
                              u"\U00002500-\U00002BEF"  # caractères chinois
                              u"\U00002702-\U000027B0"
                              u"\U00002702-\U000027B0"
                              u"\U000024C2-\U0001F251"
                              u"\U0001f926-\U0001f937"
                              u"\U00010000-\U0010ffff"
                              u"\u2640-\u2642"
                              u"\u2600-\u2B55"
                              u"\u200d"
                              u"\u23cf"
                              u"\u23e9"
                              u"\u231a"
                              u"\ufe0f"  # dingbats
                              u"\u3030"
                              "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)
    
    # 3. CARACTÈRES SPÉCIAUX : Supprimer les caractères non-ASCII
    text = re.sub(r'[^\x00-\x7F]+', ' ', text)
    
    # 4. URLs : Supprimer les URLs
    text = re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ', text)
    text = re.sub(r'www\.(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', ' ', text)
    
    # 5. MENTIONS : Supprimer les @mentions
    text = re.sub(r'@\w+', ' ', text)
    
    # 6. HASHTAGS : Supprimer les #hashtags
    text = re.sub(r'#\w+', ' ', text)
    
    # 7. CHIFFRES : Supprimer les chiffres isolés
    text = re.sub(r'\b\d+\b', ' ', text)
    
    # 8. PONCTUATION : Supprimer la ponctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # 9. ESPACES : Normaliser les espaces multiples
    text = re.sub(r'\s+', ' ', text)
    
    # 10. TRIM : Supprimer les espaces en début/fin
    text = text.strip()
    
    return text
```

```python
# ÉTAPE 3: Application du nettoyage au dataset anonymisé
print("NETTOYAGE DU DATASET ANONYMISE")
print("=" * 35)

# Vérifier que df_anonymized existe
if 'df_anonymized' not in locals():
    print("df_anonymized n'existe pas encore!")
else:
    # Travailler seulement sur les 10 000 données anonymisées
    max_to_clean = 10000
    actual_size_to_clean = min(max_to_clean, len(df_anonymized))
    
    # Créer une copie du dataset anonymisé pour le nettoyage (seulement les 10k)
    df_cleaned = df_anonymized.head(actual_size_to_clean).copy()
    
    # Statistiques de nettoyage
    total_comments = len(df_cleaned)
    original_total_words = 0
    cleaned_total_words = 0
    empty_after_cleaning = 0
    
    print(f"Nettoyage de {total_comments:,} commentaires anonymisés...")
    
    # Appliquer le nettoyage avec barre de progression
    for i in tqdm(range(total_comments), desc="Nettoyage des commentaires"):
        original_comment = df_cleaned['comment_text'].iloc[i]
        
        if original_comment and not pd.isna(original_comment):
            # Compter les mots avant nettoyage
            original_total_words += len(str(original_comment).split())
            
            # Appliquer le nettoyage standard
            cleaned_comment = clean_text(original_comment)
            
            # Remplacer dans le DataFrame
            df_cleaned.loc[i, 'comment_text'] = cleaned_comment
            
            # Compter les mots après nettoyage
            if cleaned_comment:
                cleaned_total_words += len(cleaned_comment.split())
            else:
                empty_after_cleaning += 1
    
    print(f"\nNETTOYAGE TERMINE!")
    print(f"STATISTIQUES DU NETTOYAGE:")
    print(f"Total commentaires traités: {total_comments:,}")
    print(f"Mots avant nettoyage: {original_total_words:,}")
    print(f"Mots après nettoyage: {cleaned_total_words:,}")
    if original_total_words > 0:
        print(f"Réduction: {((original_total_words - cleaned_total_words) / original_total_words * 100):.1f}%")
    print(f"Commentaires vides après nettoyage: {empty_after_cleaning}")
    
    # Supprimer les commentaires vides
    df_cleaned = df_cleaned[df_cleaned['comment_text'].str.len() > 0]
    print(f"Commentaires conservés: {len(df_cleaned):,}")
    
    # Sauvegarder le dataset nettoyé
    cleaned_filename = "dataset_cleaned_and_anonymized_10k.csv"
    df_cleaned.to_csv(cleaned_filename, index=False)
    print(f"\nDataset nettoyé sauvegardé: {cleaned_filename}")
    
    print(f"\nDATASET FINAL PRET:")
    print(f"Variable: df_cleaned")
    print(f"Anonymisé: OUI (RGPD conforme)")
    print(f"Nettoyé: OUI (prêt pour ML)")
    print(f"Taille: {len(df_cleaned):,} commentaires")
    print(f"Prêt pour l'entraînement SVM")
```

```python
# ÉTAPE 4: Analyse de la qualité du nettoyage
print("ANALYSE DE LA QUALITE DU NETTOYAGE")
print("=" * 40)

# Vérifier que les DataFrames existent
if 'df_anonymized' not in locals() or 'df_cleaned' not in locals():
    print("Veuillez d'abord exécuter les étapes d'anonymisation et de nettoyage.")
else:
    # Comparaison avant/après nettoyage
    print("COMPARAISON AVANT/APRES NETTOYAGE:")
    
    # Afficher quelques exemples de transformation
    examples_to_show = 5
    examples_count = 0
    
    for i in range(min(100, len(df_anonymized), len(df_cleaned))):
        original = df_anonymized['comment_text'].iloc[i]
        cleaned = df_cleaned['comment_text'].iloc[i]
        
        # Montrer seulement les exemples où il y a une différence significative
        if (original and cleaned and 
            str(original) != str(cleaned) and 
            len(str(original)) > 20 and 
            examples_count < examples_to_show):
            
            examples_count += 1
            print(f"\nExemple {examples_count}:")
            print(f"  AVANT: {str(original)[:120]}...")
            print(f"  APRÈS: {str(cleaned)[:120]}...")
            
            # Analyse des transformations
            orig_words = len(str(original).split())
            clean_words = len(str(cleaned).split())
            reduction = ((orig_words - clean_words) / orig_words * 100) if orig_words > 0 else 0
            
            print(f"  STATS: {orig_words} -> {clean_words} mots (-{reduction:.1f}%)")
    
    # Analyse statistique globale
    print(f"\nANALYSE STATISTIQUE GLOBALE:")
    
    # Longueur moyenne des commentaires
    if len(df_cleaned) > 0:
        original_lengths = [len(str(text).split()) for text in df_anonymized['comment_text'].head(1000) if text]
        cleaned_lengths = [len(str(text).split()) for text in df_cleaned['comment_text'].head(1000) if text]
        
        avg_original = sum(original_lengths) / len(original_lengths) if original_lengths else 0
        avg_cleaned = sum(cleaned_lengths) / len(cleaned_lengths) if cleaned_lengths else 0
        
        print(f"Longueur moyenne avant: {avg_original:.1f} mots")
        print(f"Longueur moyenne après: {avg_cleaned:.1f} mots")
        if avg_original > 0:
            print(f"Réduction moyenne: {((avg_original - avg_cleaned) / avg_original * 100):.1f}%")
    
    # Distribution des labels dans le dataset nettoyé
    if 'toxic' in df_cleaned.columns:
        toxic_distribution = df_cleaned['toxic'].value_counts()
        print(f"\nDISTRIBUTION DES LABELS (dataset nettoyé):")
        print(f"Non-toxiques: {toxic_distribution.get(0, 0):,} ({toxic_distribution.get(0, 0)/len(df_cleaned)*100:.1f}%)")
        print(f"Toxiques: {toxic_distribution.get(1, 0):,} ({toxic_distribution.get(1, 0)/len(df_cleaned)*100:.1f}%)")
    
    if 'insult' in df_cleaned.columns:
        insult_distribution = df_cleaned['insult'].value_counts()
        print(f"Non-insultants: {insult_distribution.get(0, 0):,} ({insult_distribution.get(0, 0)/len(df_cleaned)*100:.1f}%)")
        print(f"Insultants: {insult_distribution.get(1, 0):,} ({insult_distribution.get(1, 0)/len(df_cleaned)*100:.1f}%)")
    
    # Vérification de la qualité
    print(f"\nVERIFICATION DE QUALITE:")
    very_short_comments = len(df_cleaned[df_cleaned['comment_text'].str.len() < 10])
    print(f"Commentaires très courts (< 10 char): {very_short_comments}")
    print(f"Pourcentage de commentaires courts: {very_short_comments/len(df_cleaned)*100:.2f}%")
    
    print(f"\nDATASET FINAL PRET:")
    print(f"{len(df_cleaned):,} commentaires nettoyés et anonymisés")
    print(f"Conformité RGPD: Assurée")
    print(f"Nettoyage: Optimisé pour ML")
    print(f"Prêt pour l'entraînement SVM")
    
    # Aperçu du dataset final
    print(f"\nAPERCU DU DATASET FINAL:")
    display_columns = ['comment_text', 'toxic']
    if 'insult' in df_cleaned.columns:
        display_columns.append('insult')
    
    print(df_cleaned[display_columns].head(3))
    
    print(f"\nFICHIERS CREES:")
    print(f"dataset_anonymized_rgpd_10k.csv (dataset anonymisé)")
    print(f"dataset_cleaned_and_anonymized_10k.csv (dataset final)")
    print(f"Variables disponibles: df_anonymized, df_cleaned")
```
